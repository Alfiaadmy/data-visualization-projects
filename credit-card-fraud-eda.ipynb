```python
class_counts = df['Class'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=class_counts.index, y=class_counts.values)

plt.title("Distribusi Transaksi Fraud vs Non-Fraud")
plt.xlabel("Class (0 = Non Fraud, 1 = Fraud)")
plt.ylabel("Jumlah Transaksi")

plt.show()

print(class_counts)
```

```python
plt.figure(figsize=(6,6))

plt.pie(
    class_counts,
    labels=['Non Fraud','Fraud'],
    autopct='%1.2f%%',
    colors=['skyblue','salmon']
)

plt.title("Persentase Fraud vs Non-Fraud")
plt.show()
```

```python
features_to_plot = ['Time','Amount','V1','V2','V3','V4']

for col in features_to_plot:

    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=30, kde=True)

    plt.title(f"Distribusi {col}")
    plt.xlabel(col)
    plt.ylabel("Frekuensi")

    plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Amount'])

plt.title("Boxplot Amount (Melihat Outlier)")
plt.show()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(14,10))

sns.heatmap(
    correlation_matrix,
    cmap='coolwarm',
    center=0
)

plt.title("Correlation Matrix Antar Fitur")
plt.show()
```

```python
corr_with_target = df.corr()['Class'].sort_values(ascending=False)

print("Korelasi Fitur terhadap Target Fraud:\n")
print(corr_with_target)
```

```python
plt.figure(figsize=(10,6))

corr_with_target.drop('Class').plot(kind='bar')

plt.title("Korelasi Fitur terhadap Fraud")
plt.ylabel("Nilai Korelasi")
plt.xlabel("Fitur")

plt.show()
```

```python
top_features = corr_with_target.abs().sort_values(ascending=False).head(10)

print("10 Fitur dengan Korelasi Terkuat terhadap Fraud:")
top_features
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat visualisasi untuk membuktikan pengaruh V17 terhadap Fraud
plt.figure(figsize=(8, 6))
sns.boxplot(x='Class', y='V17', data=df, palette='Set2')

plt.title("Bukti Korelasi Negatif: Distribusi V17 pada Fraud vs Non-Fraud")
plt.xlabel("Class (0: Normal, 1: Fraud)")
plt.ylabel("Nilai Fitur V17")

# Menambahkan anotasi rata-rata agar lebih jelas
mean_0 = df[df['Class'] == 0]['V17'].mean()
mean_1 = df[df['Class'] == 1]['V17'].mean()

plt.text(0, mean_0, f'Mean: {mean_0:.2f}', color='black', weight='bold', ha='center')
plt.text(1, mean_1, f'Mean: {mean_1:.2f}', color='black', weight='bold', ha='center')

plt.show()
```
